##Cargar Datos

```python
# Cargar el conjunto de datos "tips" desde seaborn
dataframe = sns.load_dataset("tips")
```

```python
# Imprimir las primeras filas del DataFrame
print("DataFrame original:")
dataframe.head(10) #Dentro del parentesis de la función puedes escribir el numero de filas que deseas mostrar
```

##Manipulación y Limpieza de Datos

```python
# Verificar y manejar valores nulos
print("\nManejo de valores nulos:")
print(dataframe.isnull().sum())  # Mostrar cantidad de valores nulos por columna
```

##Filtrar el dataframe para obtener los registros de los comensales con las siguientes caracteristicas:

*   Es Hombre
*   Tiene Mesa en zona de fumar
*   Fue a Cenar








```python
Comensales = dataframe[(dataframe['sex'] == 'Male') & (dataframe['smoker'] == 'Yes') & (dataframe['time'] == 'Dinner')]
```

```python
# Imprimir el DataFrame filtrado
print("\nDataFrame con filtro:")
Comensales
```

```python
Comensales.to_csv('Comensales_Filtrados', index=False)  # Guardar el DataFrame como un archivo CSV sin incluir el índice
```

```python
# Ruta donde se guardó el archivo CSV en Google Drive
ruta_archivo = 'Comensales_Filtrados'

# Cargar el DataFrame desde el archivo CSV en Google Drive
dataframe_filtrado = pd.read_csv(ruta_archivo)
```

```python
dataframe_filtrado
```

## Gráfico


```python
# Crear el gráfico de línea
plt.figure(figsize=(8, 6))
dataframe_filtrado.groupby('day')['total_bill'].mean().plot(kind='line', marker='o', color='green')

# Configurar el título y las etiquetas de los ejes
plt.title('Total de la Cuenta por Día')
plt.xlabel('Día')
plt.ylabel('Total de la Cuenta')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
